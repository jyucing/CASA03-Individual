# 4 Join

```python
#ITA_joined = pd.merge(ITA_1, gdf1[['ISO_A3','geometry']], left_on='ISO3', right_on='ISO_A3', how = 'left')
```

```python
ITA_joined = pd.merge(gdf1, ITA_1, left_on='ISO_A3', right_on='ISO3', how = 'left')
ITE_joined = pd.merge(gdf1, ITE_1, left_on='ISO_A3', right_on='ISO3', how = 'left')
DTT_joined = pd.merge(gdf1, DTT_1, left_on='ISO_A3', right_on='ISO3', how = 'left')
EPM_joined = pd.merge(gdf1, EPM_1, left_on='ISO_A3', right_on='ISO3', how = 'left')
```

```python
# Replace '..' and '...' to be NAN
#ITA_joined.replace(['..', '...'], pd.NA, inplace=True)
#ITE_joined.replace(['..', '...'], pd.NA, inplace=True)
#DTT_joined.replace(['..', '...'], pd.NA, inplace=True)
#EPM_joined.replace(['..', '...'], pd.NA, inplace=True)
```

```python
# Convert to int class
columns_to_convert = ['1995', '1996', '1997', '1998', '1999', '2000', '2001', 
                      '2002', '2003', '2004', '2005', '2006', '2007', '2008', 
                      '2009', '2010', '2011', '2012', '2013', '2014', '2015', 
                      '2016', '2017', '2018', '2019', '2020', '2021']
ITA_joined[columns_to_convert] = ITA_joined[columns_to_convert].apply(pd.to_numeric, errors='coerce')
ITE_joined[columns_to_convert] = ITE_joined[columns_to_convert].apply(pd.to_numeric, errors='coerce')
DTT_joined[columns_to_convert] = DTT_joined[columns_to_convert].apply(pd.to_numeric, errors='coerce')
EPM_joined[columns_to_convert] = EPM_joined[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# Convert DF to GDF
ITA_joined = gpd.GeoDataFrame(ITA_joined, geometry = 'geometry')
ITE_joined = gpd.GeoDataFrame(ITE_joined, geometry = 'geometry')
DTT_joined = gpd.GeoDataFrame(DTT_joined, geometry = 'geometry')
EPM_joined = gpd.GeoDataFrame(EPM_joined, geometry = 'geometry')
```
